```python
# tipos de dados no python  - revisão

#int
idade = 17
print("idade: ")
print(idade)
print(type(idade))

#float (ponto flutuante) valores com virgula (ponto separador de decimais )
altura = 1,60
print("altura: ")
print(altura)
print(type(altura))

#str (string) texto
nome = "Maria Ingrid"
print("nome: ")
print(nome)
print(type(nome))

#bool - booleano o que é o boleano galera (sim, não, verdade, mentira, ligado, desligado) ou seja só ha duas opções
briga = False
print("briga: ")
print(briga)
print(type(briga))

#list lista (coleção ordenada)
frutas = ["blueberry","morango","manga"]
print("frutas: ")
print(frutas)
print(type(frutas))

#tuple - dupla, como por exemplo, cordenadas
coordenadas = (52,640,24,325)
print("coordenadas: ")
print(coordenadas)
print(type(coordenadas))
#set - conjunto
numeros = {1, 2, 2, 3, 3, 4, 5, 5,}
print("numeros: ")
print(numeros)
print(type(numeros))

#dict - dicionario
pessoas = {"nome":"Nathingrid","idade":17}
print("pessoas: ")
print(pessoas)
print(type(pessoas))

#Nome Type = quando não tem nada, mas preciso informar um valor
valor = None
print("valor: ")
print(valor)
print(type(valor))
```
